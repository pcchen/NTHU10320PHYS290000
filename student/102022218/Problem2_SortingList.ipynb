## Selection sort

Following is the algorithm of the **selection sort**:

* Divide the input list into two sub-lists:
 * Sorted sub-list (Start with an empty list).
 * Un-Sorted sub-list.
* Find the smallest element in the unsorted list.
* Swap with the leftmost unsorted element.
* Move the sub-list boundaries.

## Insertion sort

Following is the algorithm of the **insertion sort**:

* Divide the input list into two sub-lists:
 * Sorted sub-list (Start with the leftmost element).
 * Un-Sorted sub-list.
* Insert the leftmost element of the unsorted sub-list into the proper position of the sorted sub-list.
* Move the sub-list boundaries

```python
from random import *

# create a random list with N elements
def random_list(N):
    i=0
    l=[]
    while i<N:
        l=l+[1]
        j=random()
        x=int(N*j)
        l[i]=x

# selection sort
def selection_sort(l):
    n=len(l)
    i=0
    j=0
    l1=[]
    l2=l
    while i<=n-1:
        x=l2[0]
        while j<=(n-1)-i:
            if x>=l2[j]:
                x=l2[j]
                k=j
            j=j+1
        l1[i]=x
        del l2[k]
        i=i+1
        j=0
    return l1
# insertion sort
def insertion_sort(l):
    n=len(l)
    i=0
    j=0
    l1=[l[0]]
    l2=l[1:]
    if l2[0]>=l1[0]:
        l1[1]=l2[0]
    else:
        l1[1]=l1[0]
        l1[0]=l2[0]
    del l2[0]
    
    while i<=n-3:
        x=l2[0]
        while j<=(n-1)-i:
            if x<=l1[j]:
                k=j
                break
            else:
                j=j+1
        l1=l1+[x]
        inverseindex=i+3
        while inverseindex>=k+1:
            l1[inverseindex]=l1[inverseindex-1]
            inverseindex=inverseindex-1
        l1[k]=x
        del l2[0]
        i=i+1
        j=0
    return l1
```

```python
# test the implementation of selection sort
N = 10
L = random_list(N)
L_sorted = sorted(L)
selection_sort(L)

for i in range(len(L)):
    if L[i] != L_sorted[i]:
        print 'oops!'    
```
