# Fourier Series

The $N$-th order Fourier series of a period function $f(x)$ of period $L$ is defined as

$$
  FS(N,x) \equiv \frac{a_0}{2} + \sum_{r=1}^{N} 
  \left[
    a_r \cos\left(\frac{2\pi r x}{L}\right) + b_r \sin\left(\frac{2\pi r x}{L}\right)
  \right],
$$
where 

$$
  a_r = \frac{2}{L} \int_{x_0}^{x_0+L} f(x) \cos\left(\frac{2\pi r x}{L} \right) dx, \; \;
  b_r = \frac{2}{L} \int_{x_0}^{x_0+L} f(x) \sin\left(\frac{2\pi r x}{L} \right) dx.
$$

 \end{cases}
$$

```python
# create your functions
def fo(t):
    if t<0:
        return -1
    else:
        return 1
    
def FS(N, t):
    i=1
    sss=0
    while i<=N:
        arr=4*sin(2*pi*i*t)
        brr=2*(2-cos(2*pi*i*t))
        sos=arr*cos(2*pi*i*t)+brr*sin(2*pi*i*t)
        sss=sss+sos
        i=i+1
    return sss
        
def ppprnt(a):
    x = linspace(-0.2,0.2, 1001)
    ylim=(-1.5,1.5)
    plot(x, FS(a, x))
    show()
    
    
    
kk=-0.2    
while kk<0.2:
    
    plot(kk, fo(kk),'ko')
    kk=kk+0.01
ylim=(-1.5,1.5)
show()
ppprnt(10)
ppprnt(50)
ppprnt(100)
```
