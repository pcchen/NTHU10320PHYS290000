```python
def f1(alpha):
    n=0
    t=0.5
    while n<=1000:
        t=alpha*t*(1-t)
        n=n+1
    return t
def f2(alpha):
    m=0
    y=f1(alpha)
    for m in range(1000):
        y=alpha*y*(1-y)
        plot(alpha,y,'k-',markersize=0.1)
    show()
alpha=linspace(1,4,300)
f2(alpha)
```
