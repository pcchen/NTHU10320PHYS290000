## Problem 1

Consider the damped oscillation
$$
  m \ddot{x} + b \dot{x} + k x =0,
$$
which can be rewritten as 
$$
  \ddot{x} + 2\beta\dot{x} + \omega^2_0 x =0,
$$
where $\beta=b/2m$ is the **damping parameter** and $\omega_0=\sqrt{k/m}$ is the characteristic angular frequency in the absence of damping. 

### Underdamping
When $\omega^2_0  > \beta^2$ the system has underdamping oscillation. Define
$$
  \omega_1 = \sqrt{\omega_0^2 -\beta^2}.
$$
The general solution can be written as 
$$
  x(t) = A \exp(-\beta t) \cos(\omega_1 t - \delta).
$$

### Overdsamping
When $\omega^2_0  < \beta^2$ the system has underdamping oscillation. Define
$$
  \omega_2 = \sqrt{\beta^2 - \omega_0^2}.
$$
The general solution can be written as
$$
  x(t) = A \exp(-\beta t) \left[ A_1 \exp(\omega_2 t) + A_2 \exp(-\omega_2 t) \right].
$$

For a given set of $m$, $b$, $k$, and 
* $A$, $\delta$ (for underdamping)
* $A_1$, $A_2$ (for overdamping)

Do the following:
* Plot $x(t)$.
* Plot $E(t)= \frac{1}{2} k x^2(t) + \frac{1}{2} m \dot{x}^2(t)$.
* Plot the $\dot{x}(t)$ v.s. $x(t)$.

```python
# Problem 1
%pylab inline
?figure

print 'm*x_doubledot + b*x_dot + k*x = 0  same as \nx_doubledot + 2B*x_dot + w0**2 *x = 0'
# determine the coefficient of damped oscillation equation
m = float(raw_input('m = '))
b = float(raw_input('b = '))
k = float(raw_input('k = '))
B  = b / (2*m)
w0 = k / m

# determine the coefficient of underdamped solution
A = float(raw_input('A = '))
delta = float(raw_input('delta = '))

# determine the coefficient of overdamped solution
A1 = float(raw_input('A1 = '))
A2 = float(raw_input('A2 = '))

def x(t):
    # Underdamping
    if w0**2 > B**2:
        w1 = sqrt(w0**2 - B**2)
        return A * e**(-B*t) * cos(w1*t - delta)
    # Overdamping
    else:
        w2 = sqrt(B**2 - w0**2)
        return A * e**(-B*t) * (A1 * e**(w2*t) + A2 * e**(-w2*t))  

def x_dot(t):
    # Underdamping
    if w0**2-B**2>0:
        w1 = sqrt(w0**2 - B**2)
        return -B * A * e**(-B*t) * cos(w1*t-delta) - A * w1 * e**(-B*t) * sin(w1*t - delta)
    # Overdamping
    else:
        w2 = sqrt(B**2 - w0**2)
        return -B * A * e**(-B*t) * (A1 * e**(w2*t) + A2 * e**(-w2*t)) + A * e**(-B*t)(w2 * A1 * e**(w2*t) - w2 * A2 * e**(-w2*t))      
    
    
def E(t):
    return 0.5 * k * x(t)**2 + 0.5 * m * x_dot(t)**2
    
t1 = linspace(0, 30, 600)
plot(t1, x(t1))
xlabel('t')
ylabel('x(t)')
show()
    
t2 = linspace(0, 30, 600)
plot(t2, x_dot(t2))
xlabel('t')
ylabel('x_dot(t)')
show()

t3 = linspace(0, 30, 600)
plot(x(t3), x_dot(t3))
xlabel('x(t)')
ylabel('x_dot(t)')
show()

t4 = linspace(0, 10, 200)
plot(t4, E(t4))
xlabel('t')
ylabel('E(t)')
show()
```
